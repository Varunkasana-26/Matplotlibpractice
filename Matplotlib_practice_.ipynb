```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
x = np.arange(0,10)
y = np.arange(11,21)
```

```python
a = np.arange(40,50)
b = np.arange(50,60)
```

```python
plt.scatter(x,y,c='r')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph in 2D')
plt.savefig('graph.png')
```

```python
plt.plot( x,y,'ro-')
```

```python
plt.plot( x,y,'r--', linestyle='dashed', linewidth=5,markersize=10)
```

```python
plt.subplot(3,1,2)
plt.plot(x,y,'r--')
plt.subplot(3,1,1)
plt.plot(a,b)
plt.subplot(3,1,3)
plt.plot(a,b,'g')
```

```python
x = np.arange(0,4*np.pi,0.1)
y = np.sin(x)
plt.title('sine wave form')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.plot(x,y)
```

```python
x =np.arange(0,4*np.pi,0.1)
y = np.cos(x)
plt.title('cos wave form')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.plot(x,y)
```

```python
x = np.arange(0,4*np.pi,0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)
plt.subplot(2,1,1)
plt.plot(x,y_sin,'r--')
plt.title('sine')
plt.subplot(2,1,2)
plt.plot(x,y_cos,'g--')
plt.title('cos')
```
